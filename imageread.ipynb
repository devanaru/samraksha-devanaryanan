```python
from IPython.display import Image
treader = easyocr.Reader(['en'],gpu=False)
out = treader.readtext('adhar.PNG')
user = 'Nilesh Singh'
for i in out:
    if user in i:
        cord = i
print(cord)
result=cord[0:1][0:1]
result=result[0]
print(result)
import matplotlib.pyplot as plt
```

```python
min_x,min_y= [int(min(idx)) for idx in zip(*result)]
max_x,max_y=[int(max(idx)) for idx in zip(*result)]
image=cv2.imread('adhar.PNG')
cv2.rectangle(image,(min_x,min_y),(max_x,max_y),(0,0,255),2)
plt.figure(figsize=(8,18))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
birth = 'birth.PNG'
Image.open(birth)
```

```python
birthcert = easyocr.Reader(['en'],gpu=False)
final = birthcert.readtext(birth)
final
```

```python
child = 'Vaishali'
father = 'Jaydevbhai'
mother = 'Renukaben'
father2=''
for i in range(len(father)):
    if(father[i]=='v'):
        father2=father2+'r'
    else:
        father2=father2+father[i]
dad = ''
mom =''
son = ''
for i in final:
    if child in i[1][0:]:
        son = i
    if (father in i[1][0:] or father2 in i):
         dad = i
    if mother in i[1][0:] or mother in i[1][0:]:
         mom = i
print(son)
print(dad)
print(mom)
```

```python
tn=son[0:1][0:1]
mom=mom[0:1][0:1]
dad=dad[0:1][0:1]
tn=tn[0]
mom=mom[0]
dad=dad[0]
print(tn)
```

```python
min_x,min_y=[int(min(idx)) for idx in zip(*tn)]
max_x,max_y=[int(max(idx)) for idx in zip(*tn)]
dad_minx,dad_miny=[int(min(idx)) for idx in zip(*dad)]
dad_maxx,dad_maxy=[int(max(idx)) for idx in zip(*dad)]
mom_minx,mom_miny=[int(min(idx)) for idx in zip(*mom)]
mom_maxx,mom_maxy=[int(max(idx)) for idx in zip(*mom)]
image=cv2.imread('birth.PNG')
cv2.rectangle(image,(min_x,min_y),(max_x,max_y),(0,0,255),2)
cv2.rectangle(image,(dad_minx,dad_miny),(dad_maxx,dad_maxy),(0,0,255),2)
cv2.rectangle(image,(mom_minx,mom_miny),(mom_maxx,mom_maxy),(0,0,255),2)
plt.figure(figsize=(18,18))
plt.axis('off')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
